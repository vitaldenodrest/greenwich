```python
import sympy as sp
sp.init_printing()

import IPython.display as disp
```

```python
H2, chi, gam = sp.symbols('P^2 chi gamma')

P2_H2 = H2 - chi * H2**2
disp.display(P2_H2)
```

```python
P2, alpha0, alpha1, alpha2, alpha3, alpha4, alpha5 = sp.symbols('H^2, alpha0, alpha1, alpha2, alpha3, alpha4, alpha5')

H2_P2 = alpha1*P2 + alpha2*(P2**2)
disp.display(H2_P2)
```

```python
P2_P2 = P2_H2.subs(H2, H2_P2)
disp.display(P2_P2)
```

```python
P2_P2 = P2_P2.expand()
disp.display(P2_P2)
```

```python
P2_P2 = sp.Poly(P2_P2, P2)
coeff = []
for i in range(8):
    disp.display(P2_P2.coeff_monomial(P2**i))
    coeff.append(P2_P2.coeff_monomial(P2**i))
```

```python
sp.solve([#coeff[0],
          coeff[1] - 1,
          coeff[2],
          #coeff[3],
          #coeff[4],
          ],
         (alpha1, alpha2))
```
